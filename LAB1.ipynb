### 2.9.2 через `DataFrame.apply`

Используем `df.apply` чтобы решить следующую задачу: столбец **Доходы, всего** является объектным типом данных. Нам он нужен как `float`. Применим `df.apply` ко всему столбцу, перед этим написав функцию удаления пробела из строки. 

```python
def cut_space(s):
    s = "".join(s.split())
    return s
```

```python
data['Доходы, всего'] = data['Доходы, всего'].apply(cut_space)
data['Доходы, всего'] = data['Доходы, всего'].apply(float)
```
